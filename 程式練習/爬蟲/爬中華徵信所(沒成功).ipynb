```python
#用javascript執行點擊
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import time

# 指定Edge WebDriver的路径
PATH = r"C:\selenium\msedgedriver.exe"
service = Service(PATH)

# 设置下载路径
download_path = r"C:\Users\user\OneDrive\Desktop\匯率資料"
edge_options = webdriver.EdgeOptions()
edge_options.add_experimental_option("prefs", {
    "download.default_directory": download_path,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
})

# 初始化Edge WebDriver
driver = webdriver.Edge(service=service, options=edge_options)

# 打开网页
driver.get('https://dbquery.credit.com.tw')

def login(usercode, username, password):
    try:
        # 等待登录框加载完成
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "TextBox_UserCode"))
        )
        
        # 输入用户信息
        driver.find_element(By.ID, "TextBox_UserCode").send_keys(usercode)
        driver.find_element(By.ID, "TextBox_UserName").send_keys(username)
        driver.find_element(By.ID, "TextBox_Password").send_keys(password)
        
        # 等待登录按钮变为可点击状态
        login_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, "submit"))
        )
        
        # 使用 JavaScript 点击按钮
        driver.execute_script("arguments[0].click();", login_button)
        
        # 等待登录成功后的页面元素加载完成
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "some_element_id"))  # 替换为登录成功后的页面元素
        )
        print("登录成功")
    except TimeoutException:
        print("登录超时，请检查网络或页面元素")
    except NoSuchElementException:
        print("未找到页面元素，请检查ID是否正确")
    except Exception as e:
        print(f"登录过程中发生错误: {e}")

# 调用登录函数
usercode = "S0000926"
username = "nccu"
password = "nccu0201"
login(usercode, username, password)

# 等待5秒
time.sleep(5)

# 关闭浏览器
driver.quit()
```
